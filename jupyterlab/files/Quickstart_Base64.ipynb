```python
# use pandas for data frame manipulation
import pandas as pd

# read the 'quickstart_base64.csv' file
base64_cells = pd.read_csv("quickstart_base64.csv")
```

```python
# inspect raw data (show table)
base64_cells
```

```python
# create new columns to facilitate analysis

# IMPLEMENTATION ID (CODE UNIT + ADAPTER) 
base64_cells['IMPLID'] = base64_cells['SYSTEMID'] + '_' + base64_cells['ADAPTERID']
# STATEMENT ID (SHEETS AND COORDINATES)
base64_cells['STATEMENT'] = base64_cells['SHEETID'].str.rsplit(pat='_', n=1, expand=True)[0] + '@' + base64_cells['X'].astype(str) + ',' + base64_cells['Y'].astype(str)
```

```python
# let's look at the output columns (where type == 'value') of all actuation sheets
base64_outputs = base64_cells.query('TYPE == "value" & ARENAID == "execute" & SHEETID != "all"').sort_values(by=['STATEMENT'])
base64_outputs
```

```python
# SRH-like operations
# here we use pivoting on the output columns of actuation sheets to verify functional correctness

# first drop redudant pseudo oracles of implementations (to avoid duplicate entries)
base64_outputs.drop_duplicates(subset=['IMPLID', 'STATEMENT', 'VALUE'], inplace=True)

# pivot the data to allow column-wise comparisons
base64_outputs.pivot(index='STATEMENT', columns= 'IMPLID', values= 'VALUE').sort_values(by=['STATEMENT'])
```

```python
# next to behavioural correctness, let's analyze code coverage measurements

# assume that we indicate the complexity of implementations by the number of branches executed (i.e., jacoco_class_branch.covered)

base64_cc_branch = base64_cells.query('TYPE == "jacoco_class_branch.covered" & ARENAID == "jacoco"')
base64_cc_branch
```
